## Mini game: guess Py's favorite number

Py is thinking of a number between 1 and 10. Type a guess and click the
button!

```python
import random
import ipywidgets as widgets
from IPython.display import display

secret_number = random.randint(1, 10)
guess_box = widgets.IntText(value=1, description="Guess:")
guess_button = widgets.Button(description="Check my guess!")
guess_output = widgets.Output()


def on_guess(_):
    with guess_output:
        guess_output.clear_output()
        if guess_box.value == secret_number:
            celebrate(f"Yes! Py's number was {secret_number}!")
        elif guess_box.value > secret_number:
            try_again(hint="Py's number is smaller.")
        else:
            try_again(hint="Py's number is bigger.")


guess_button.on_click(on_guess)
display(guess_box, guess_button, guess_output)
```

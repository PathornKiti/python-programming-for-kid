## How animation works

A game "loop" does the same thing many times per second:

1. Move things a tiny bit (the apple drops a few pixels).
2. Redraw the whole picture.

Do that fast enough and your eyes see smooth movement! The **▶ play
button** below is a timer that runs our `tick()` function again and again.

Run the cell, press **▶**, and catch some apples with the buttons!

````python
from ipycanvas import Canvas, hold_canvas
import ipywidgets as widgets
from IPython.display import display
import random

WIDTH, HEIGHT = 400, 400
BASKET_W = 90        # how wide your basket is
FRUIT_R = 15         # how big the fruit is
FALL_SPEED = 10      # how many pixels the fruit drops each frame

canvas = Canvas(width=WIDTH, height=HEIGHT)

score = 0
misses = 0
basket_x = WIDTH // 2 - BASKET_W // 2
fruit_x = random.randint(FRUIT_R, WIDTH - FRUIT_R)
fruit_y = 0
status = widgets.Output()


def draw():
    with hold_canvas(canvas):
        canvas.clear()
        # sky
        canvas.fill_style = "#bde0fe"
        canvas.fill_rect(0, 0, WIDTH, HEIGHT)
        # the falling apple
        canvas.fill_style = "#e63946"
        canvas.fill_circle(fruit_x, fruit_y, FRUIT_R)
        # the basket
        canvas.fill_style = "#6d4c41"
        canvas.fill_rect(basket_x, HEIGHT - 30, BASKET_W, 18)


def show_status(message=""):
    with status:
        status.clear_output()
        print(f"🍎 Caught: {score}    😅 Missed: {misses}")
        if message:
            print(message)


def new_fruit():
    global fruit_x, fruit_y
    fruit_x = random.randint(FRUIT_R, WIDTH - FRUIT_R)
    fruit_y = 0


def tick(change):
    global fruit_y, score, misses
    fruit_y = fruit_y + FALL_SPEED
    if fruit_y >= HEIGHT - 30:
        if basket_x <= fruit_x <= basket_x + BASKET_W:
            score = score + 1
        else:
            misses = misses + 1
        new_fruit()
        show_status()
    draw()


def move_left(_):
    global basket_x
    basket_x = max(0, basket_x - 35)
    draw()


def move_right(_):
    global basket_x
    basket_x = min(WIDTH - BASKET_W, basket_x + 35)
    draw()


play = widgets.Play(value=0, min=0, max=1000000, step=1, interval=120)
play.observe(tick, names="value")

left_button = widgets.Button(description="⬅️ Left")
right_button = widgets.Button(description="➡️ Right")
left_button.on_click(move_left)
right_button.on_click(move_right)

new_fruit()
draw()
show_status("Press the ▶ play button, then use Left / Right to catch the apples!")
display(canvas)
display(widgets.HBox([left_button, play, right_button]))
display(status)
````

## Read the important part

The heart of the game is the `tick()` function. Every frame it does:

```python
fruit_y = fruit_y + FALL_SPEED     # move the apple down
if fruit_y >= HEIGHT - 30:         # did it reach the basket line?
    if basket_x <= fruit_x <= basket_x + BASKET_W:
        score = score + 1         # basket was under it -> caught!
    else:
        misses = misses + 1       # basket missed it
    new_fruit()                    # drop a fresh apple from the top
```

That's just the loops and `if`/`else` you already learned — now making a
game move!

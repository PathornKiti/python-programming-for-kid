## Your turn: tell Py your name

Type your name in the box, then click the button.

```python
import ipywidgets as widgets
from IPython.display import display

name_box = widgets.Text(placeholder="Type your name here")
say_hi_button = widgets.Button(description="Say hi to Py!")
greeting_output = widgets.Output()


def on_say_hi(_):
    with greeting_output:
        greeting_output.clear_output()
        name = name_box.value.strip() or "friend"
        celebrate(f"Hi {name}! Let's learn Python together! 🐢")


say_hi_button.on_click(on_say_hi)
display(name_box, say_hi_button, greeting_output)
```

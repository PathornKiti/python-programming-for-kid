## Storing a value

A variable is a labeled box that holds a value. Run this:

```python
stars_drawn = 5
print(stars_drawn)
```

Now Py remembers the number `5` under the name `stars_drawn`. We can
reuse it: 

```python
stars_drawn = stars_drawn + 1
print("Py has now drawn", stars_drawn, "stars!")
```

## Variables can hold text too

```python
turtle_name = "Py"
print("My turtle's name is", turtle_name)
```

## Doing math

Python can add `+`, subtract `-`, multiply `*`, and divide `/`.

```python
length = 4
width = 3
area = length * width
print("The rectangle's area is", area)
```

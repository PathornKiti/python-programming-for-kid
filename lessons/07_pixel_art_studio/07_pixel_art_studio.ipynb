## How this works

`ipycanvas` gives Python a picture you can click on. Every click runs
**your code**, which decides what happens — in this case, coloring in
one square of a grid.

Run the cell below, then click anywhere on the white grid to paint it!

```python
from ipycanvas import Canvas
import ipywidgets as widgets
from IPython.display import display

CELL_SIZE = 25
GRID_SIZE = 16

canvas = Canvas(width=GRID_SIZE * CELL_SIZE, height=GRID_SIZE * CELL_SIZE)


def draw_grid_lines():
    canvas.stroke_style = "#dddddd"
    for i in range(GRID_SIZE + 1):
        canvas.stroke_line(i * CELL_SIZE, 0, i * CELL_SIZE, canvas.height)
        canvas.stroke_line(0, i * CELL_SIZE, canvas.width, i * CELL_SIZE)


def clear_canvas():
    canvas.fill_style = "white"
    canvas.fill_rect(0, 0, canvas.width, canvas.height)
    draw_grid_lines()


current_color = {"value": "#e63946"}


def on_click(x, y):
    col = int(x // CELL_SIZE)
    row = int(y // CELL_SIZE)
    canvas.fill_style = current_color["value"]
    canvas.fill_rect(col * CELL_SIZE, row * CELL_SIZE, CELL_SIZE, CELL_SIZE)


canvas.on_mouse_down(on_click)
clear_canvas()
display(canvas)
```

## Pick a color

Click a swatch below, then go click on your picture above to paint with
that color.

```python
PALETTE = [
    ("Red", "#e63946"),
    ("Orange", "#f4a261"),
    ("Yellow", "#f1c40f"),
    ("Green", "#2a9d8f"),
    ("Blue", "#457b9d"),
    ("Purple", "#8e44ad"),
    ("Black", "#1d1d1d"),
    ("Eraser", "white"),
]

swatch_buttons = []


def make_color_handler(hex_value):
    def handler(_):
        current_color["value"] = hex_value
    return handler


for name, hex_value in PALETTE:
    btn = widgets.Button(description=name)
    btn.style.button_color = hex_value
    btn.on_click(make_color_handler(hex_value))
    swatch_buttons.append(btn)

clear_button = widgets.Button(description="🗑️ Clear", button_style="danger")
clear_button.on_click(lambda _: clear_canvas())

display(widgets.HBox(swatch_buttons))
display(clear_button)
```

## Bonus: Rainbow Splash

Here's a fun trick — a loop that paints 40 random squares in random
colors, all in one click!

```python
import random

RAINBOW = ["#e63946", "#f4a261", "#f1c40f", "#2a9d8f", "#457b9d", "#8e44ad"]


def rainbow_splash(_):
    for _ in range(40):
        col = random.randint(0, GRID_SIZE - 1)
        row = random.randint(0, GRID_SIZE - 1)
        canvas.fill_style = random.choice(RAINBOW)
        canvas.fill_rect(col * CELL_SIZE, row * CELL_SIZE, CELL_SIZE, CELL_SIZE)


splash_button = widgets.Button(description="🌈 Rainbow Splash!")
splash_button.on_click(rainbow_splash)
display(splash_button)
```

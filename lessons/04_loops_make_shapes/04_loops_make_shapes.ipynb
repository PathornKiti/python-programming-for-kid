## The `for` loop

`for i in range(4):` means "do the next lines **4 times**." Compare this
to lesson 3 — same square, way less typing:

```python
clearscreen()
initializeTurtle()

for i in range(4):
    forward(100)
    right(90)
```

## Your turn: draw a hexagon

A hexagon has **6** sides. Each turn should be **60** degrees (a full
turtle turn is always 360 degrees, so 360 / 6 = 60). Change the numbers
below:

```python
clearscreen()
initializeTurtle()
pencolor("blue")

for i in range(6):
    forward(80)
    right(60)
```

## Drawing a star

Stars use a bigger turn — try 144 degrees!

```python
clearscreen()
initializeTurtle()
pencolor("gold")
width(2)

for i in range(5):
    forward(150)
    right(144)
```
